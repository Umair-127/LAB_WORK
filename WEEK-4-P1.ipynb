WEEK4_Q2 - Write a function to check whether a given string is a palindrome.

```python
def is_palindrome(text):
    text = text.replace(" ", "").lower()

    if text == text[::-1]:
        return True
    else:
        return False

word = input("Enter a string: ")

if is_palindrome(word):
    print("It is a Palindrome.")
else:
    print("It is not a Palindrome.")
```

WEEK4_Q3 - Create a dictionary of student names and marks. Display students scoring above 75.

```python
students = {
    "Alice": 82,
    "Bob": 67,
    "Charlie": 91,
    "David": 74,
    "Emma": 88
}

print("Students scoring above 75:")

for name, marks in students.items():
    if marks > 75:
        print(name, ":", marks)
```

WEEK4_Q4 - Write a program to sort a list in ascending and descending order.

```python
numbers = []
n=int(input("Enter the num of elements = "))

for i in range(n):
    num=int(input(f"Enter {i} number = "))
    numbers.append(num)

ascending=[]
descending=[]

for i in range(len(numbers)):
  for j in range(i + 1, len(numbers)):
    if numbers[i] > numbers[j]:
      numbers[i], numbers[j] = numbers[j], numbers[i]
  ascending.append(numbers[i])


for i in range(len(numbers)):
  for j in range(i + 1, len(numbers)):
    if numbers[i] < numbers[j]:
      numbers[i], numbers[j] = numbers[j], numbers[i]
  descending.append(numbers[i])

print("Sorted list:", ascending)
print("Descending :",descending)
```
