WEEK3_Q1 - Write a Python program to read a number and check whether it is even or odd.


```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is Even")
else:
    print("The number is Odd")
```

WEEK3_Q2 - Write a Python program to find the largest of three numbers using conditional statements.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print("The largest number is:", largest)
```

WEEK3_Q3 - Write a Python program to generate the first n Fibonacci numbers using a loop.

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1

print("Fibonacci Series:")

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

WEEK3_Q4 - Write a Python program to count the number of vowels in a given string.

```python
string = input("Enter a string: ")
count = 0

for char in string:
    if char.lower() in "aeiou":
        count += 1

print("Number of vowels:", count)
```
